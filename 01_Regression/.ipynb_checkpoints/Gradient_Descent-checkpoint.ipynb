```python
import numpy as np
import matplotlib.pyplot as plt
```

**Gradient Descent**

```python
# Hypothesis --> H_w(x) = (w0 + w1 * x)
def h_w(x, w):
    return w[0] + w[1] * x


# Cost Function (SSE here)
def cost_function():
    return np.sum((h_w(x, w) - y)**2) / len(x)


def gradient_descent(x, y, W, steps, alpha):
    n = len(x)
    
    for step in range(steps):
        grad_w0 = np.sum(h_w(x,W) - y) / n
        grad_w1 = np.sum((h_w(x,W) - y) * x) / n
        W[0] -= alpha * grad_w0
        W[1] -= alpha * grad_w1

    return W
```

Generate random data

```python
def gen_data(count=500, noise=10, slope=6, intercept=8):
    x = np.random.normal(20, 1.5, count)
    e = np.random.randn(count) * noise
    y = slope * x + intercept + e
    return x, y
```

```python
n = 500
x, y = gen_data(count=n, noise=10, slope=6, intercept=8)
plt.scatter(x, y, label="y = f(x)")
plt.title("Randomly Generated Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
w_init = [0, 0]
w_final = gradient_descent(x, y, w_init, steps=500, alpha=0.1)
y_hat = w_final[0] + w_final[1] * x

# Visualize the predicted line
plt.scatter(x, y, label="y = f(x)")
plt.plot(x, y_hat, color='red')
plt.title(f"Predicted Line (y = {w_final[1]}x {"+" if (w_final[0] >= 0) else "-"} {np.abs(w_final[0]):.2f})")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
